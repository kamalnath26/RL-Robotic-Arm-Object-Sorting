```python
from stable_baselines3 import DDPG
import gymnasium as gym
import numpy as np
import panda_gym
import matplotlib.pyplot as plt

# Load the environment and model
env = gym.make("PandaPickAndPlace-v3", render_mode="human")
model = DDPG.load("ddpg_Panda_Arm_pick_place_rgb_arr", env=env)

vec_env = model.get_env()

# ✅ Evaluation function to collect episode rewards
def evaluate_and_plot(model, vec_env, n_eval_episodes=10):
    rewards_per_episode = []

    for episode in range(n_eval_episodes):
        obs = vec_env.reset()
        total_reward = 0.0
        done = [False]

        while not done[0]:
            action, _states = model.predict(obs, deterministic=True)
            obs, reward, done, info = vec_env.step(action)
            total_reward += reward[0]  # reward is a vector in DummyVecEnv

        rewards_per_episode.append(total_reward)
        print(f"Episode {episode + 1}: Reward = {total_reward:.2f}")

    # ✅ Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(rewards_per_episode, marker='o', label='Episode Rewards')
    plt.title('Pretrained Model Evaluation')
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# ✅ Run evaluation and show plot
evaluate_and_plot(model, vec_env, n_eval_episodes=1000)
```
